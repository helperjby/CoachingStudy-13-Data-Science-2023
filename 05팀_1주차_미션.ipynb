## Q1. 여러분은 파이썬을 통해 설문조사 문항의 응답내역을 분석하게 되었습니다.
문항별 응답내용에는 하나의 응답만 할 수 있는 single choice 문제와 여러 응답을 선택할 수 있는 multiple choice 문제가 있습니다. 2개를 구분하기 위해 single choice 문항 번호에 "_"를 표기하지 않기로 했습니다.

문항별 응답내역이 'question'에 담겨 있을 때, 조건문과 반복문을 사용하여 아래와 같은 결과가 출력되도록 코드를 작성해보세요.


```python
# 응답 설문 문항
question = ['Q2', 'Q3', 'Q4', 'Q5', 'Q6_1', 'Q6_2', 'Q6_3', 'Q6_4', 'Q6_5', 'Q6_6',
       'Q6_7', 'Q6_8', 'Q6_9', 'Q6_10', 'Q6_11', 'Q6_12', 'Q7_1', 'Q7_2',
       'Q7_3', 'Q7_4', 'Q7_5', 'Q7_6', 'Q7_7', 'Q8', 'Q9', 'Q10_1', 'Q10_2',
       'Q10_3']
```

### A1

### 장병용

```python
sc = []
# mc = []
for i in question:
  if len(i) < 4:
    sc.append(i)

print(sc)
```

### 민소희

```python
# for(반복문)을 이용하여 빈 배열에 값이 있는지 확인 후 입력하는 방식
arr = []
for i in question:
    str = i.split("_")[0]
    if(str not in arr):
        arr.append(str)


print(arr)
```

```python
#  or(반복문)을 이용하여 배열을 다시 만들어주고 중복을 제거하는방법
arr = []
for i in question:
    arr.append(i.split("_")[0])

list(dict.fromkeys(arr))
```

### 이준환

```python
original_list = ['Q2', 'Q3', 'Q4', 'Q5', 'Q8', 'Q9']
modified_list = [f'Q6_{i}' for i in range(1, 13)] + [f'Q7_{i}' for i in range(1, 8)] + original_list + [f'Q10_{i}' for i in range(1, 4) ]

print("question =", modified_list)
```

만약 행에는 '연도'가, 열에 '국가'가 들어있고 수치 데이터의 결과값이 아래와 같다면, 출력형태는 조금 달라도 괜찮습니다.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

### 장병용

```python
# 연도별-국가별 평균 기대수명
df_2011 = df[df['Year'] > 2010]
grouped_df_2011 = df_2011.groupby(['Year', 'Country'])['Life_Expectancy'].mean().reset_index()
year_country_avg_life = grouped_df_2011.set_index(['Year','Country']).unstack()
year_country_avg_life
```
